# Statistiques descriptives

## Variable: Elimination du gaspillage

```python
mapping_dict = {
    "Tout à fait d’accord":6,
    "D’accord":5,
    "Légèrement d’accord":4,
    "Légèrement pas d’accord":3,
    "Pas d’accord":2,
    "Pas du tout d'accord":1,
}

index_1 = ["count","mean","std","median", "mode"]
drop_index_1 = ["min","25%","75%","max"]
```

### Statistiques descriptives de la sous-variable: Agilité du processus


```python
A_P = df.iloc[:,1:5]
A_P
```

```python
A_P = A_P.replace(mapping_dict)
A_P
```

```python
A_P_Items = A_P.columns.to_list()
A_P_Items
```

```python
A_P_Stats = A_P.describe()
mode = A_P.mode()
A_P_Stats.drop(drop_index_1,inplace=True)
A_P_Stats = pd.concat([A_P_Stats, mode], axis=0)
A_P_Stats.index = index_1
A_P_Stats
```

```python
A_P_mean = A_P.mean(axis=1).to_frame()
A_P_mean.plot.hist()
plt.xlabel("Agilité du processus", fontsize=10)
plt.xticks(fontsize=8)
plt.ylabel("Fréquences", labelpad=20)
plt.title("Fréquences de la sous-variable agilité du processus")

plt.ylim(0, 40 * 1.2) 
plt.show()
```

```python
A_P_mean.max()
```

```python
mode = A_P_mean.mode()
A_P_mean_stats = A_P_mean.describe()
A_P_mean_stats.drop(drop_index_1,inplace=True)
A_P_mean_stats = pd.concat([A_P_mean_stats,mode], axis=0)
# A_P.columns = ["Statistics"]
A_P_mean_stats.index = index_1
A_P_mean_stats.rename(columns={0:"Statistics"}, inplace=True)
A_P_mean_stats
```

#### Test de validité

#### Test de fiabilité

```python
alpha = pg.cronbach_alpha(data=A_P)[0]
alpha
```
